# Gaussian distribution

```python
%matplotlib inline

import matplotlib.pyplot as plt
import torch

from model.gauss import Gaussian

plt.style.use("seaborn-darkgrid")
```

# Data

```python
x_dim = 1
loc = 1.4
scale = 1.2

n = 100

x = scale * torch.randn(n, x_dim) * scale + loc
```

```python
plt.hist(x.numpy(), bins=20)
plt.show()
```

# Model

```python
gauss = Gaussian(x_dim)
```

```python
gauss.inference(x)

print(gauss.mu, gauss.Sigma)
```

```python
x_axis = torch.range(-3, 6, 0.1)
pdf_log = gauss.get_log_prob({"x": x_axis})
pdf = torch.exp(pdf_log)
```

```python
plt.hist(x.numpy(), bins=20, density=True)
plt.plot(x_axis, pdf)
plt.show()
```

# Sample

```python
sample = gauss.sample(sample_shape=[100])

plt.hist(sample["x"].numpy(), bins=20, density=True)
plt.plot(x_axis, pdf)
plt.show()
```
